```python
# Import necessary libraries
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()

# Convert to a Pandas DataFrame for easier handling
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add the target column
iris_df['species'] = iris.target

# Map target values to species names
iris_df['species'] = iris_df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Display the first 5 rows
print("First five rows of the Iris dataset:")
print(iris_df.head())
```

```python
# Randomly sample 30 observations
random_sample = iris_df.sample(n=30, random_state=42)

print("Randomly sampled 30 observations:")
print(random_sample)
```

```python
# Sample Mean Distribution Analysis
import numpy as np
import matplotlib.pyplot as plt

# Store sample means
sample_means = []

# Perform random sampling 100 times
for _ in range(100):
    sample = iris_df.sample(n=30)
    sample_mean = sample['sepal length (cm)'].mean()
    sample_means.append(sample_mean)

# Plot histogram of sample means
plt.hist(sample_means, bins=10, color='skyblue', edgecolor='black')
plt.title("Distribution of 100 Sample Means of Sepal Length")
plt.xlabel("Sample Mean of Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
# Define a function for systematic sampling
def systematic_sampling(data, step):
    indices = np.arange(0, len(data), step)
    return data.iloc[indices]

# Perform systematic sampling (20% of the dataset)
step = len(iris_df) // 5  # For 20%, use step size of 1/5th the length
sampled_data = systematic_sampling(iris_df, step)

print("Systematically sampled data (20% of the dataset):")
print(sampled_data)

# Plot the original and sampled datasets
plt.figure(figsize=(12, 6))

# Original dataset
plt.subplot(1, 2, 1)
plt.scatter(range(len(iris_df)), iris_df['sepal length (cm)'], color='blue', label='Original')
plt.title("Original Dataset")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.legend()

# Sampled dataset
plt.subplot(1, 2, 2)
plt.scatter(range(len(sampled_data)), sampled_data['sepal length (cm)'], color='red', label='Sampled')
plt.title("Sampled Dataset")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.legend()

plt.tight_layout()
plt.show()
```
